```python
# â”€â”€ SETUP â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€
!pip install groq -q

from groq import Groq
from google.colab import userdata

client = Groq(api_key=userdata.get('GROQ_API_KEY'))

def call_ai(system_prompt, user_message):
    response = client.chat.completions.create(
        model="llama-3.3-70b-versatile",
        messages=[
            {"role": "system", "content": system_prompt},
            {"role": "user",   "content": user_message}
        ],
        temperature=0.3
    )
    return response.choices[0].message.content.strip()

# â”€â”€ STEP 0: GET USER INPUT â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€
print("=" * 55)
print("       CUSTOMER SUPPORT AI â€” PROMPT CHAIN")
print("=" * 55)
customer_message = input("\nğŸ’¬ Describe your issue: ")

# â”€â”€ STEP 1: CLASSIFY â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€
print("\n" + "=" * 55)
print("STEP 1 â€” CLASSIFYING YOUR ISSUE...")
print("=" * 55)

step1_system = """
You are a customer support triage assistant.
Classify the customer message and respond ONLY in this exact format:
Category: <Shipping, Billing, Account, Technical, Returns, or Other>
Urgency: <Low, Medium, or High>
Sentiment: <Neutral, Frustrated, or Angry>
Summary: <one sentence describing the issue>

Rules:
- Use High urgency if customer says urgent, ASAP, immediately, or has a deadline.
- Use Angry if customer says unacceptable, furious, terrible, worst.
- Do not include anything outside the four lines above.
"""

step1_output = call_ai(step1_system, customer_message)
print("\n" + step1_output)

# Parse Step 1 output
lines = step1_output.strip().split("\n")
classification = {}
for line in lines:
    if ":" in line:
        key, val = line.split(":", 1)
        classification[key.strip()] = val.strip()

category  = classification.get("Category", "Other")
urgency   = classification.get("Urgency", "Low")
sentiment = classification.get("Sentiment", "Neutral")

# â”€â”€ STEP 2: CLARIFYING QUESTIONS (uses Step 1 output) â”€â”€â”€â”€â”€â”€â”€â”€
print("\n" + "=" * 55)
print("STEP 2 â€” GATHERING MORE INFORMATION...")
print("=" * 55)

step2_system = f"""
You are a friendly customer support assistant.
The customer has a {category} issue with {urgency} urgency and seems {sentiment}.

Ask EXACTLY 3 short clarifying questions to gather info needed to resolve a {category} issue.
Number each question (1. 2. 3.).
Tone: warm, empathetic, professional.
Do NOT attempt to solve the problem yet.
Do NOT ask more than 3 questions.
"""

step2_output = call_ai(step2_system, customer_message)
print("\n" + step2_output)

print("\nğŸ’¬ Please answer the questions above:")
customer_followup = input("Your answers: ")

# â”€â”€ STEP 3: PROPOSE SOLUTION (uses Step 1 + Step 2 + answers)â”€
print("\n" + "=" * 55)
print("STEP 3 â€” GENERATING YOUR SOLUTION...")
print("=" * 55)

step3_system = f"""
You are an expert customer support agent for an e-commerce company.
You have classified this as a {category} issue with {urgency} urgency.
Customer sentiment is {sentiment}.

Using the original message, clarifying questions, and customer answers, write a response that:
- Acknowledges frustration if sentiment is Frustrated or Angry
- Gives a clear solution with concrete next steps
- Stays under 100 words
- Tone: warm, professional, solution-focused
- Do NOT make up policies â€” use standard e-commerce practices
"""

step3_input = f"""
Original message: {customer_message}
Clarifying questions asked: {step2_output}
Customer answers: {customer_followup}
"""

step3_output = call_ai(step3_system, step3_input)
print("\n" + step3_output)

# â”€â”€ STEP 4: ESCALATION CHECK (uses Step 1 + Step 3) â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€
print("\n" + "=" * 55)
print("STEP 4 â€” ESCALATION DECISION...")
print("=" * 55)

step4_system = f"""
You are a customer support supervisor.
Decide if this ticket should be ESCALATED to a human or RESOLVED by AI.

Escalate if ANY are true:
- Sentiment is Angry
- Urgency is High AND category is Billing
- Urgency is High AND category is Shipping
- Customer asked to speak to a human

Respond ONLY in this exact format:
Decision: <ESCALATE or RESOLVE>
Reason: <one sentence>
Priority: <Urgent, Priority, or Standard â€” only if ESCALATE>
"""

step4_input = f"""
Classification: {step1_output}
Proposed solution: {step3_output}
Original message: {customer_message}
"""

step4_output = call_ai(step4_system, step4_input)
print("\n" + step4_output)

# â”€â”€ FINAL SUMMARY â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€
print("\n" + "=" * 55)
print("              TICKET SUMMARY")
print("=" * 55)
print(f"  Issue    : {customer_message[:60]}...")
print(f"  Category : {category}")
print(f"  Urgency  : {urgency}")
print(f"  Sentiment: {sentiment}")
print(f"\n  Solution : {step3_output[:80]}...")
print(f"\n  {step4_output}")
print("\nâœ… Prompt chain complete â€” all 4 steps ran successfully!")
```

```python
# â”€â”€ SETUP â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€
!pip install groq -q

from groq import Groq
from google.colab import userdata

client = Groq(api_key=userdata.get('GROQ_API_KEY'))

def call_ai(system_prompt, user_message):
    response = client.chat.completions.create(
        model="llama-3.3-70b-versatile",
        messages=[
            {"role": "system", "content": system_prompt},
            {"role": "user",   "content": user_message}
        ],
        temperature=0.3
    )
    return response.choices[0].message.content.strip()

# â”€â”€ WELCOME â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€
print("=" * 55)
print("       CUSTOMER SUPPORT AI â€” PROMPT CHAIN")
print("=" * 55)
print("Type 'quit' at any time to exit.\n")
customer_message = input("ğŸ’¬ Hi! What can I help you with today?\nYou: ")

if customer_message.lower() == "quit":
    print("Goodbye! Have a great day!")
else:

    # â”€â”€ STEP 1: CLASSIFY â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€
    print("\n" + "=" * 55)
    print("STEP 1 â€” CLASSIFYING YOUR ISSUE...")
    print("=" * 55)

    step1_system = """
    You are a customer support triage assistant.
    Classify the customer message and respond ONLY in this exact format:
    Category: <Shipping, Billing, Account, Technical, Returns, or Other>
    Urgency: <Low, Medium, or High>
    Sentiment: <Neutral, Frustrated, or Angry>
    Summary: <one sentence describing the issue>

    Rules:
    - Use High urgency if customer says urgent, ASAP, immediately, or has a deadline.
    - Use Angry if customer says unacceptable, furious, terrible, worst.
    - Do not include anything outside the four lines above.
    """

    step1_output = call_ai(step1_system, customer_message)
    print("\n" + step1_output)

    lines = step1_output.strip().split("\n")
    classification = {}
    for line in lines:
        if ":" in line:
            key, val = line.split(":", 1)
            classification[key.strip()] = val.strip()

    category  = classification.get("Category", "Other")
    urgency   = classification.get("Urgency", "Low")
    sentiment = classification.get("Sentiment", "Neutral")

    # â”€â”€ STEP 2: CLARIFYING QUESTIONS â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€
    print("\n" + "=" * 55)
    print("STEP 2 â€” GATHERING MORE INFORMATION...")
    print("=" * 55)

    step2_system = f"""
    You are a friendly customer support assistant.
    The customer has a {category} issue with {urgency} urgency and seems {sentiment}.
    Ask EXACTLY 3 short clarifying questions to gather info needed to resolve a {category} issue.
    Number each question (1. 2. 3.).
    Tone: warm, empathetic, professional.
    Do NOT attempt to solve the problem yet.
    Do NOT ask more than 3 questions.
    """

    step2_output = call_ai(step2_system, customer_message)
    print("\nğŸ¤– Agent:", step2_output)

    print("\nğŸ’¬ Please answer the questions above:")
    customer_followup = input("You: ")

    # â”€â”€ STEP 3: PROPOSE SOLUTION â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€
    print("\n" + "=" * 55)
    print("STEP 3 â€” GENERATING YOUR SOLUTION...")
    print("=" * 55)

    step3_system = f"""
    You are an expert customer support agent for an e-commerce company.
    You have classified this as a {category} issue with {urgency} urgency.
    Customer sentiment is {sentiment}.
    Using the original message, clarifying questions, and customer answers, write a response that:
    - Acknowledges frustration if sentiment is Frustrated or Angry
    - Gives a clear solution with concrete next steps
    - Stays under 100 words
    - Tone: warm, professional, solution-focused
    - Do NOT make up policies â€” use standard e-commerce practices
    """

    step3_input = f"""
    Original message: {customer_message}
    Clarifying questions asked: {step2_output}
    Customer answers: {customer_followup}
    """

    step3_output = call_ai(step3_system, step3_input)
    print("\nğŸ¤– Agent:", step3_output)

    # â”€â”€ STEP 4: ESCALATION CHECK â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€
    print("\n" + "=" * 55)
    print("STEP 4 â€” ESCALATION DECISION...")
    print("=" * 55)

    step4_system = f"""
    You are a customer support supervisor.
    Decide if this ticket should be ESCALATED to a human or RESOLVED by AI.
    Escalate if ANY are true:
    - Sentiment is Angry
    - Urgency is High AND category is Billing
    - Urgency is High AND category is Shipping
    - Customer asked to speak to a human
    Respond ONLY in this exact format:
    Decision: <ESCALATE or RESOLVE>
    Reason: <one sentence>
    Priority: <Urgent, Priority, or Standard â€” only if ESCALATE>
    """

    step4_input = f"""
    Classification: {step1_output}
    Proposed solution: {step3_output}
    Original message: {customer_message}
    """

    step4_output = call_ai(step4_system, step4_input)
    print("\n" + step4_output)

    escalation_lines = step4_output.strip().split("\n")
    escalation = {}
    for line in escalation_lines:
        if ":" in line:
            key, val = line.split(":", 1)
            escalation[key.strip()] = val.strip()

    # â”€â”€ STEP 5: RESOLUTION CHECK LOOP â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€
    print("\n" + "=" * 55)
    print("STEP 5 â€” CHECKING IF ISSUE IS RESOLVED...")
    print("=" * 55)

    max_attempts = 3
    attempt = 0
    issue_resolved = False

    while attempt < max_attempts and not issue_resolved:
        attempt += 1
        print(f"\nğŸ¤– Agent: Has your issue been resolved? (attempt {attempt} of {max_attempts})")
        resolved_input = input("You: ").lower()

        step5_system = f"""
        You are a customer support agent wrapping up a support conversation.
        The customer's original issue was: {category} - {classification.get('Summary', '')}
        The solution provided was: {step3_output}

        The customer just said: "{resolved_input}"

        Determine if the customer is satisfied or still needs help.
        If satisfied: respond warmly, thank them, and close the ticket.
        If not satisfied: apologize sincerely and ask ONE specific follow-up question
        to better understand what is still wrong.
        Keep response under 60 words.
        Tone: warm, empathetic, professional.
        """

        step5_output = call_ai(step5_system, resolved_input)
        print(f"\nğŸ¤– Agent: {step5_output}")

        # Check if resolved based on customer's words
        positive_words = ["yes", "yeah", "yep", "resolved", "fixed", "thanks", "thank you", "great", "perfect", "good", "works", "worked", "ok", "okay"]
        negative_words = ["no", "nope", "not", "still", "unresolved", "problem", "issue", "help", "broken", "doesn't", "cant", "can't"]

        if any(word in resolved_input for word in positive_words):
            issue_resolved = True
        elif any(word in resolved_input for word in negative_words):
            # Give one more attempt at a solution
            print("\n" + "=" * 55)
            print("STEP 6 â€” PROVIDING ADDITIONAL HELP...")
            print("=" * 55)

            followup_response = input("\nğŸ’¬ Tell us more about what's still not working:\nYou: ")

            step6_system = f"""
            You are an expert customer support agent.
            The customer's issue ({category}) was not fully resolved by the first solution.
            Original solution given: {step3_output}
            Customer says it's still not fixed: {followup_response}

            Provide an alternative solution or escalation path.
            Be specific and actionable.
            Keep under 80 words.
            Tone: apologetic, professional, solution-focused.
            """

            step6_output = call_ai(step6_system, followup_response)
            print(f"\nğŸ¤– Agent: {step6_output}")

    # â”€â”€ STEP 7: ESCALATION IF STILL UNRESOLVED â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€
    if not issue_resolved and attempt >= max_attempts:
        print("\n" + "=" * 55)
        print("STEP 7 â€” ESCALATING TO HUMAN AGENT...")
        print("=" * 55)
        print("\nğŸ¤– Agent: I'm so sorry we haven't been able to fully resolve your issue.")
        print("   I'm escalating this to a senior support agent who will contact")
        print("   you within 2 hours via email. Your ticket has been marked URGENT.")
        escalation["Decision"] = "ESCALATE"
        escalation["Priority"] = "Urgent"

    # â”€â”€ FINAL SUMMARY â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€
    print("\n" + "=" * 55)
    print("              TICKET SUMMARY")
    print("=" * 55)
    print(f"  Issue    : {customer_message[:60]}...")
    print(f"  Category : {category}")
    print(f"  Urgency  : {urgency}")
    print(f"  Sentiment: {sentiment}")
    print(f"  Resolved : {'âœ… Yes' if issue_resolved else 'âŒ No â€” Escalated'}")
    print(f"  Decision : {escalation.get('Decision', 'RESOLVE')}", end="")
    if escalation.get('Priority'):
        print(f" ({escalation.get('Priority')})", end="")
    print(f"\n  Reason   : {escalation.get('Reason', 'N/A')}")
    print("\nâœ… Prompt chain complete â€” all steps ran successfully!")
```

Exercise 1 â€” Prompt Chaining for a Customer Support AI
Tool used: Google Colab + Groq API (LLaMA 3.3 model)

The 1st block of code is the 1st verison the 2nd is the improved model.



Prompts used:

Step 1 â€” Classify the Issue
The AI acts as a triage assistant and classifies the customer's message into a category, urgency level, and sentiment. This output is used by every step that follows.

Step 2 â€” Gather Missing Information
The AI uses the category and sentiment from Step 1 to ask exactly 3 relevant clarifying questions. The tone is kept warm and empathetic and it is instructed not to attempt a solution yet.

Step 3 â€” Propose a Solution
Using the classification from Step 1, the questions from Step 2, and the customer's answers, the AI generates a solution under 100 words with clear next steps.

Step 4 â€” Escalation Decision
The AI acts as a supervisor and uses Step 1's urgency and sentiment to decide whether to escalate to a human agent or resolve automatically.

Step 5 â€” Resolution Check
The AI asks the customer if their issue was resolved. If not it apologizes and asks one follow-up question. This loops up to 3 times.

Step 6 â€” Follow-up Solution
If still unresolved the AI uses the original solution from Step 3 and the customer's new explanation to generate an alternative fix.

Step 7 â€” Auto Escalation
If unresolved after all attempts the system automatically escalates the ticket to a human agent and marks it urgent. No AI prompt needed â€” handled by logic.
