````python
# â”€â”€ SETUP â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€
!pip install groq -q

from groq import Groq
from google.colab import userdata
import re

client = Groq(api_key=userdata.get('GROQ_API_KEY'))

def call_ai(system_prompt, user_message):
    response = client.chat.completions.create(
        model="llama-3.3-70b-versatile",
        messages=[
            {"role": "system", "content": system_prompt},
            {"role": "user",   "content": user_message}
        ],
        temperature=0.3
    )
    return response.choices[0].message.content.strip()

# â”€â”€ TASK â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€
task = "Write a Python program that sorts and filters a list of student grades, shows passing students from highest to lowest, calculates class statistics, and handles invalid inputs."

print("=" * 50)
print("   ReACT AGENT â€” STUDENT GRADE SORTER")
print("=" * 50)

# â”€â”€ STAGE 1: REASON â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€
print("\nğŸ“Œ STAGE 1: REASONING...")

stage1_output = call_ai("""
You are a Python developer using ReACT (Reason + Act).
STAGE: REASON â€” Think through the task before writing code.
Respond in this format:
THOUGHT: <what does this task need?>
PLAN: <3 bullet points on how to build it>
EDGE CASES: <what could go wrong?>
Do NOT write any code yet.
""", task)

print(stage1_output)

# â”€â”€ STAGE 2: GENERATE CODE â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€
print("\nğŸ“Œ STAGE 2: GENERATING CODE...")

stage2_output = call_ai("""
You are a Python developer using ReACT (Reason + Act).
STAGE: ACT â€” Write the Python code based on the plan.
Requirements:
- Use only built-in Python libraries
- Include 8 sample students with varied grades (some failing below 60)
- Filter failing students, sort passing students high to low
- Show class average, highest, and lowest grade
- Handle empty lists and invalid inputs
- Keep code clean with comments

Wrap code in ```python ``` blocks.
""", f"Task: {task}\n\nPlan:\n{stage1_output}")

print(stage2_output)

# Extract code
code_match = re.search(r'```python(.*?)```', stage2_output, re.DOTALL)
extracted_code = code_match.group(1).strip() if code_match else ""

# â”€â”€ STAGE 3: RUN & OBSERVE â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€
print("\nğŸ“Œ STAGE 3: RUNNING CODE...")
print("-" * 50)

try:
    exec(extracted_code)
    execution_result = "SUCCESS"
    print("-" * 50)
    print("âœ… Code ran successfully!")
except Exception as e:
    execution_result = f"ERROR: {e}"
    print(f"âŒ {execution_result}")

# â”€â”€ STAGE 4: OBSERVE â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€
print("\nğŸ“Œ STAGE 4: OBSERVING RESULTS...")

stage4_output = call_ai("""
You are a Python developer using ReACT (Reason + Act).
STAGE: OBSERVE â€” Review the code and execution result.
Respond in this format:
OBSERVATION: <did it work?>
ISSUES: <any problems found or say None>
NEXT: <say FIX if issues, DONE if everything passed>
Keep it brief â€” 3 lines max.
""", f"Code:\n{extracted_code}\n\nResult: {execution_result}")

print(stage4_output)

# â”€â”€ STAGE 5: FIX â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€
print("\nğŸ“Œ STAGE 5: FIXING & IMPROVING...")

stage5_output = call_ai("""
You are a Python developer using ReACT (Reason + Act).
STAGE: FIX â€” Improve the code based on observations.
- Fix any bugs found
- Make sure all requirements are met
- Keep it clean and readable
List fixes made, then wrap improved code in ```python ``` blocks.
""", f"Original code:\n{extracted_code}\n\nObservations:\n{stage4_output}")

print(stage5_output)

improved_match = re.search(r'```python(.*?)```', stage5_output, re.DOTALL)
improved_code = improved_match.group(1).strip() if improved_match else extracted_code

# â”€â”€ STAGE 6: FINAL RUN â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€
print("\nğŸ“Œ STAGE 6: FINAL CODE OUTPUT...")
print("-" * 50)

try:
    exec(improved_code)
    print("-" * 50)
    print("âœ… Final code ran successfully!")
    final_result = "SUCCESS"
except Exception as e:
    print(f"âŒ Error: {e}")
    final_result = "FAILED"

# â”€â”€ SUMMARY â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€
print("\n" + "=" * 50)
print("         ReACT CYCLE COMPLETE")
print("=" * 50)
print("  Stage 1 â€” REASON  : Planned the approach")
print("  Stage 2 â€” ACT     : Generated the code")
print("  Stage 3 â€” RUN     : Executed and tested")
print("  Stage 4 â€” OBSERVE : Reviewed the results")
print("  Stage 5 â€” FIX     : Improved the code")
print("  Stage 6 â€” FINAL   : Ran the improved code")
print(f"\n  Result: {final_result}")
print("=" * 50)
````

# Exercise 2 â€” ReACT Prompting for Student Grade Sorter

## Tool Used
Google Colab + Groq API (LLaMA 3.3 model)

## My Prompts By Step

**Stage 1 â€” Reason**
"Before writing any code, think through what needs to be done. What does this program need to do, what is the plan to build it, and what could break or go wrong? Don't write any code yet."

**Stage 2 â€” Act (Generate Code)**
"Now write the actual Python code based on the plan. Use only built in Python, include 8 students with different grades, filter out anyone failing below 60, sort passing students highest to lowest, show class average, highest and lowest grade, and handle weird inputs like empty lists."

**Stage 3 â€” Run**
"Execute the generated code and capture whether it ran successfully or threw an error."

**Stage 4 â€” Observe**
"Check if the code worked properly. Did it run ok, are there any issues, and does it need to be fixed or is it done?"

**Stage 5 â€” Fix**
"Fix and improve the code based on what we found. Fix any bugs, make sure it does everything from the original task, and tell me what you changed."

**Stage 6 â€” Final Run**
"Run the improved code one final time to confirm everything works."

## How Each Step Uses Prior Output
- Stage 2 uses the plan from Stage 1
- Stage 4 uses the code from Stage 2 and the execution result from Stage 3
- Stage 5 uses the code from Stage 2 and the observations from Stage 4
- Stage 6 runs the fixed code from Stage 5


The model that I created uses the ReACT prompting in order to learn or check if any mistakes where made. The 1st reponse checks if it checked the students grade correctly if not try to improve on it.
